```python
import numpy as np
import pandas as pd
import os
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import f1_score
```

```python
data = pd.read_csv('DSL-StrongPasswordData.csv')
data = data.replace('s002', 1)
data = data.replace('s003', 0)
data_2_3 = data.head(800)
data_2_3 = data_2_3.sample(frac=1)
y_data = data_2_3['subject']
x_data = data_2_3.drop('subject', 1)
x_data = x_data.drop('sessionIndex', 1)
x_data = x_data.drop('rep', 1)
seed = 1
(x_train, x_val, y_train, y_val) = train_test_split(x_data, y_data, test_size=0.1,
            random_state=seed)
y_train=y_train.astype('int')
```

```python
x_train
```

```python
def add_layer(inputs, in_size, out_size, activation_function=None):
    
    if in_size>20:
        inputs = tf.nn.l2_normalize(inputs,dim=1)

    Weights = tf.Variable(tf.random_normal([in_size, out_size]))
    biases = tf.Variable(tf.random_normal([1, out_size]))
    Wx_plus_b = tf.matmul(inputs, Weights) + biases
    if activation_function is None:
        outputs = Wx_plus_b
    else:
        outputs = activation_function(Wx_plus_b)
    return outputs
```

```python
xs = tf.placeholder(tf.float32)
ys = tf.placeholder(tf.float32)
l1 = add_layer(xs, 31, 50, activation_function = tf.nn.relu)
l2 = add_layer(l1, 50, 50, activation_function = tf.nn.relu) 
l3 = add_layer(l2, 50, 30, activation_function = tf.nn.relu)
l4 = add_layer(l3, 30, 30, activation_function = tf.nn.relu)
l5 = add_layer(l4, 30, 30, activation_function = tf.nn.relu)
l6 = add_layer(l5, 30, 30, activation_function = tf.nn.relu)
l7 = add_layer(l6, 30, 20, activation_function = tf.nn.relu)
l8 = add_layer(l7, 20, 20, activation_function = tf.nn.relu)
l9 = add_layer(l8, 20, 10, activation_function = tf.nn.relu)
prediction = add_layer(l9, 10, 1, activation_function = None)
loss = tf.reduce_mean(tf.reduce_sum(tf.square(ys - prediction),
                                   reduction_indices = [1]))
# cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels = prediction,logits =  ys))
# cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=prediction,labels=ys))
train_step = tf.train.MomentumOptimizer(0.00001,0.9).minimize(loss)
init = tf.global_variables_initializer()
```

```python
sess = tf.Session()
sess.run(init)
for i in range(10000):
    sess.run(train_step, feed_dict = {xs:x_train, ys:y_train})
    if i%100 ==0:
        print(sess.run(loss, feed_dict={xs:x_train, ys:y_train}))
```
